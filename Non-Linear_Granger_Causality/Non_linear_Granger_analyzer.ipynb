```python
import pandas as pd
```

```python
file_name="mass_analyzer_Twitter_non_causality_var_negative_Sentiment"
# Read CSV file with multi-level index and header
df = pd.read_csv(file_name+'.csv', index_col=[0, 1])

# Check for duplicate index values and drop rows
df = df[~df.index.duplicated(keep='first')]

# If you want to reset index after dropping duplicate rows
df.reset_index(inplace=True)
print(df.columns.tolist())
```

```python
df
```

```python
columns_to_count = ['p_value_1', 'p_value_2', 'p_value_3','p_value_4', 'p_value_5', 'p_value_6','p_value_7', 'p_value_8', 'p_value_9','p_value_10']
# Dictionary to store count results for each column
count_results = {}

# Loop through each column and count values greater than the threshold
for column in columns_to_count:
    filtered_values = df[column] <= 0.01
    count_results[column] = filtered_values.sum()/500* 100

# Create a new DataFrame with count results
count_df = pd.DataFrame(count_results, index=[0])

new_column_names = [f"Lag_{i+1}" for i in range(10)]
count_df.columns = new_column_names
count_df = count_df.rename_axis('Proportion of stocks which have significant non-linear Granger causality')
```

```python
with open(file_name+'.tex', 'w') as f:
    f.write(count_df.to_latex(index=False))
```
